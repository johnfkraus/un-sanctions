```python

ent_link_list = []

for i in range(df_ent_norm.shape[0]):
    source = df_ent_norm['REFERENCE_NUMBER'].loc[i]

    if (df_ent_norm['TARGETS'].loc[i] is not None) and (len(df_ent_norm['TARGETS'].loc[i]) > 0):        
        temp_targets_list = []

        for target in df_ent_norm['TARGETS'].loc[i]:
            # target = target.replace('(','').replace(')','')
            if source == target:
                msg = '[WARNING] ' + source + ' equals ' + target
                print(msg)
                df_ent_norm.at[i, 'WARNING'] = msg # temp_targets_list
                # raise Exception(msg)
                continue                
            temp_targets_list.append(target)
            # name = df_ent_norm['NAME'].loc[i] 
            link2 = [df_ent_norm['REFERENCE_NUMBER'].loc[i], target, df_ent_norm['NAME'].loc[i]]
            ent_link_list.append(link2)
            df_ent_norm.at[i, 'TARGETS'] = temp_targets_list

# print(df_ent_norm['TARGETS'].value_counts())
t = df_ent_norm['TARGETS']
```
